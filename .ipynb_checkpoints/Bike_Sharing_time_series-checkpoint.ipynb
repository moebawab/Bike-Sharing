```python
import pandas as pd 
import numpy as np

# Visualization packages
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"hour.csv")
```

```python
df.info()
```

```python
#type cast datetime columns to date
df["dteday"] = pd.to_datetime(df["dteday"])
```

```python
df.head()
```

```python
df.describe(include='all').T
```

```python
df2 = pd.read_csv(r"day.csv")
```

```python
df2.info()
```

```python
#type cast datetime columns to date
df2["dteday"] = pd.to_datetime(df2["dteday"])
```

```python
df2.describe(include='all').T
```

#Grouping 

```python
# Group the DataFrame by date and time using daily frequency
df_day = df.groupby(pd.Grouper(key='dteday'))
```

```python
df_day['cnt'].sum()
```

## Visualizations

```python
# Note: Some of the Visualizations shown below are made for the dataset 'Hour' and 'day'
```

```python
#Numerical Variables

sns.pairplot(df[["temp","atemp","hum","windspeed","casual","registered","cnt"]])
plt.show()
```

```python
#Continuous Variables 

plt.figure(figsize=(20, 12))
plt.subplot(2,3,1)
plt.boxplot(df["temp"])
plt.subplot(2,3,2)
plt.boxplot(df["atemp"])
plt.subplot(2,3,3)
plt.boxplot(df["hum"])
plt.subplot(2,3,4)
plt.boxplot(df["windspeed"])
plt.subplot(2,3,5)
plt.boxplot(df["casual"])
plt.subplot(2,3,6)
plt.boxplot(df["registered"])
plt.show()
```

```python
#Plotting box plot of categorical variables

plt.figure(figsize=(20, 12))
plt.subplot(3,3,1)
sns.boxplot(x = 'season', y = 'cnt', data = df)
plt.subplot(3,3,2)
sns.boxplot(x = 'yr', y = 'cnt', data = df)
plt.subplot(3,3,3)
sns.boxplot(x = 'mnth', y = 'cnt', data = df)
plt.subplot(3,3,4)
sns.boxplot(x = 'holiday', y = 'cnt', data = df)
plt.subplot(3,3,5)
sns.boxplot(x = 'weekday', y = 'cnt', data = df)
plt.subplot(3,3,6)
sns.boxplot(x = 'workingday', y = 'cnt', data = df)
plt.subplot(3,3,7)
sns.boxplot(x = 'weathersit', y = 'cnt', data = df)
plt.show()
```
